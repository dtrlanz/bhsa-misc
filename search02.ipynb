# Phrases
I'm looking for ways to search for sequences of specific words.
This page seems a good place to start: https://annotation.github.io/text-fabric/tf/about/searchusage.html

For starters, I'll try finding instances of הנה immediately followed by a qatal verb (20 results only).

```python
A.table(A.search('''
sentence
    hn:word lex_utf8=הנה
    vb:word sp=verb vt=perf
hn <: vb
''')[0:20])
```

# Clauses
Next I'd like to search for adjacent clauses. 

## Two adjacent clauses
E.g., a clause containing הנה followed by a clause containing a participle.

```python
A.table(A.search('''
sentence
    c1:clause
        word lex_utf8=הנה
    c2:clause
        word vt=ptca
c1 <: c2
''')[0:20])
```

## Clause without verbs
Now the same search but without any verbs in the first clause.

```python
A.table(A.search('''
sentence
    c1:clause
    /without/
        word sp=verb
    /-/
        word lex_utf8=הנה
    c2:clause
        word vt=ptca
c1 <: c2
'''))
```

# Word order
Search for several words in a clause that occur in a specific order but are not necessarily adjacent.

E.g.
- a finite verb
- an adjective after the verb
- a participle after the verb
- regardless of order between adjective and participle

```python
A.table(A.search('''
clause
    vb:word vt=perf|impf|wayq
    adj:word sp=adjv
    ptc:word vt=ptcp
vb < adj
vb < ptc
''')[0:10]) #show 10 results only
```

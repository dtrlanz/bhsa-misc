# Verbless interrogative clauses with ל
- specifically yes/no questions (with הֲ)

## Except הֲלֹא
- except הֲלֹא rhetorical questions, since those are very common

```python
A.table(A.search('''
clause
/without/
    word sp=verb
/-/
/without/
    word lex_utf8=ה sp=inrg
    <: word lex_utf8=לא
/-/
    word lex_utf8=ה sp=inrg
    word lex_utf8=ל sp=prep
'''))
```

## Starting with הֲלֹא
- now just those starting with הֲלֹא 

```python
A.table(A.search('''
clause
/without/
    word sp=verb
/-/
    word lex_utf8=ה sp=inrg
    <: word lex_utf8=לא
    word lex_utf8=ל sp=prep
'''))
```

# Idem per idem
This is a preliminary attempt to find item-per-idem constructions. The search is too broad but hopefully catches all relevant occurrences.

```python
A.table(A.search('''
sentence
    clause
        a:word sp=verb
    <: clause
        word lex_utf8=אשׁר
        < b:word sp=verb vt=impf
a .lex_utf8. b
a .ps. b
a .nu. b
a .gn. b
'''))
```

Notable results: Exod 3:14; 33:19 (x2); Ezek 12:25

# Verbal clauses with מִדֶּ֫רֶךְ
I'm interested in the phrase מִדֶּ֫רֶךְ, specifically the verbs it is used with. I'll start off with a simple search for the phrase:

## Simple phrase search

```python
A.table(A.search('''
clause
    word lex_utf8=מן
    <: word lex_utf8=דרך
'''))
```

## Verbs
Now I'd like a list of all the verbs מִדֶּ֫רֶךְ is used with.

```python
A.table(A.search('''
l:lex sp=verb
/with/
clause
    w:word
    word lex_utf8=מן
    <: word lex_utf8=דרך
l .lex. w
/-/
'''))
```

It looks like there are repetitions in the above list, but that appears to be because it includes Aramaic forms. I'm only interested in the Hebrew.

## Verbs (Hebrew only)

```python
A.table(A.search('''
l:lex sp=verb language=Hebrew
/with/
clause
    w:word
    word lex_utf8=מן
    <: word lex_utf8=דרך
l .lex. w
/-/
'''))
```
